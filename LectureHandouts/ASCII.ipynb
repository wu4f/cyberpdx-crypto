```python
from langchain.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o", open_ai_key="")

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a helpful assistant that translates hex-encoded bytes into ASCII characters.  If you receive input that is not hex-encoded, return the error 'Sorry, I only work with hex-encoded input'"
        ),
        ("human", "{message}")
    ]
)

chain = prompt | llm
```

```python
result = chain.invoke({"message" : """41 20 6d 69 73 74 61 6b 65 20 73 68 6f 75 6c 64 20 62 65 20
                                      79 6f 75 72 20 74 65 61 63 68 65 72 2c 20 6e 6f 74 20 79 6f 75 72 20 
                                      61 74 74 61 63 6b 65 72 20 2d 20 41 6e 6f 6e 79 6d 6f 75 73"""})
print(result.content)
```
